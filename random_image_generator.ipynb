```
import requests
from io import BytesIO
import random
import time
from PIL import Image as PILImage
import matplotlib.pyplot as plt
import numpy as np


def plot_image_from_url(image_url, code, source_name="Wikipedia/Commons"):
    """Fetches an image from a URL and plots it using Matplotlib."""
    try:
        headers = {
            'User-Agent': 'MyRandomImagePlotter/1.0 (https://example.com/my_script; my_email@example.com) PythonRequests/2.x',
            'Cache-Control': 'no-cache',
            'Pragma': 'no-cache'
        }

        response = requests.get(image_url, headers=headers, allow_redirects=True, timeout=20)
        response.raise_for_status()

        pil_img = PILImage.open(BytesIO(response.content))
        pil_img = pil_img.convert('RGB')  # Convert to RGB for consistent plotting
        img_array = np.array(pil_img)

        plt.figure(figsize=(8, 7))
        plt.imshow(img_array)
        plt.title(f"Image Code: {code}\nSource: {source_name}")
        plt.axis('off')
        plt.show()  # Corrected indentation here

        print(f"Successfully plotted image {code} from: {response.url}")
        return True

    except Exception as e:
        print(f"Error loading or plotting image {code} from {image_url}: {e}")  # Print the error message
        # Additional handling if needed, e.g., display a placeholder image
        return False

def plot_image_result(image_url, code, source_name="Wikipedia/Commons"):
    """Fetches an image from a URL and plots it using Matplotlib."""
    try:
        headers = {
            'User-Agent': 'MyRandomImagePlotter/1.0 (https://example.com/my_script; my_email@example.com) PythonRequests/2.x',
            'Cache-Control': 'no-cache',
            'Pragma': 'no-cache'
        }

        response = requests.get(image_url, headers=headers, allow_redirects=True, timeout=20)
        response.raise_for_status()

        pil_img = PILImage.open(BytesIO(response.content))
        pil_img = pil_img.convert('RGB')  # Convert to RGB for consistent plotting
        img_array = np.array(pil_img)

        plt.figure(figsize=(8, 7))
        plt.imshow(img_array)
        plt.title(f"Image Code: {code}\nSource: {source_name}")
        plt.axis('off')
        #plt.show()  # Corrected indentation here

        print(f"Successfully plotted image {code} from: {response.url}")
        return True

    except Exception as e:
        print(f"Error loading or plotting image {code} from {image_url}: {e}")  # Print the error message
        # Additional handling if needed, e.g., display a placeholder image
        return False

# ... (rest of the code remains the same) ...


def get_random_wikimedia_commons_image_url(excluded_urls=set(), max_retries=10):
    session = requests.Session()
    session.headers.update({
        'User-Agent': 'MyRandomImagePlotter/1.0 (https://example.com/my_script; my_email@example.com) PythonRequests/2.x'
    })

    api_url = "https://commons.wikimedia.org/w/api.php"  # Corrected indentation here

    for attempt in range(max_retries):  # Corrected indentation here
        params_random = {
            "action": "query",
            "format": "json",
            "list": "random",
            "rnnamespace": "6",
            "rnlimit": "5"
        }

        try:
            response_random = session.get(url=api_url, params=params_random, timeout=10)
            response_random.raise_for_status()
            random_files = response_random.json()["query"]["random"]

            for random_file in random_files:
                title = random_file["title"].replace("File:", "")
                image_url = f"https://commons.wikimedia.org/wiki/Special:FilePath/{title}"  # Corrected indentation here
                if image_url not in excluded_urls:
                    return image_url
            time.sleep(1)
        except Exception as e:
            print(f"Error getting random image URL (attempt {attempt + 1}/{max_retries}): {e}")
            time.sleep(1)

    return None


def esperar_tecla_enter():
    """Espera até que o usuário pressione a tecla Enter."""
    input("Pressione Enter para continuar...")
    print("Tecla Enter pressionada. Continuando execução.")


def generate_code():
    return ''.join(random.choices("0123456789", k=6))


# --- Gerar 2 imagens com 2 códigos únicos ---
used_urls = set()
image_data = []  # Store codes and URLs

for _ in range(2):
    image_url = get_random_wikimedia_commons_image_url(excluded_urls=used_urls)
    if image_url:
        used_urls.add(image_url)
        code = generate_code()
        image_data.append((code, image_url))  # Store code and URL
    else:
        print("Não foi possível obter uma imagem válida.")

# Check for errors or no data before displaying images
if not image_data:
    print("Nenhuma imagem válida foi encontrada.")
else:
    # Display the codes first
    print("Generated Codes:")
    for code, _ in image_data:
        print(code)
    for code, image_url in image_data:
        plot_image_result(image_url, code)

    esperar_tecla_enter()  # Wait for Enter key press

    # Then, display the images
    for code, image_url in image_data:
        plot_image_from_url(image_url, code)
```
